```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img1 = cv2.imread('Picture1.png', cv2.IMREAD_GRAYSCALE).astype(np.float32) 
img2 = cv2.imread('Picture2.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img3 = cv2.imread('Picture3.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img5 = cv2.imread('Picture5.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)

img1 = cv2.resize(img1, (img3.shape[1], img3.shape[0]))
img2 = cv2.resize(img2, (img3.shape[1], img3.shape[0]))
img5 = cv2.resize(img5, (img3.shape[1], img3.shape[0]))
```

```python
def graytone_function(img1,M):
    return M-img1

def get_g(img1,M):
    g = graytone_function(img1,M)-img1
    return g

def log_add(img_a, img_b, M):
    a = get_g(img_a,M)
    b = get_g(img_b,M)
    added = a + b- ((a * b)/M)
    added = np.clip(added, 0, 255)
    return added

def log_subtract(img_a, img_b, M):
    a = get_g(img_a,M)
    b = get_g(img_b,M)
    subtracted = M * (a - b) / (M - b)
    subtracted = np.clip(subtracted, 0, 255)
    return subtracted

def log_multiply(img_a, img_b, M, eps=1e-12):
    a = get_g(img_a,M)
    b = get_g(img_b,M)
    phi_a = -M * np.log(np.clip(1 - (a / M), eps, 1))
    phi_b = -M * np.log(np.clip(1 - (b / M), eps, 1))
    product = phi_a * phi_b
    phi_inverse = M * (1 - np.exp(-product / M))
    return np.clip(phi_inverse, 0, M)

def log_divide(img_a, img_b, M, eps=1e-12):
    a = get_g(img_a,M)
    b = get_g(img_b,M)
    phi_a = -M * np.log(np.clip(1 - (a / M), eps, 1))
    phi_b = -M * np.log(np.clip(1 - (b / M), eps, 1))
    quotient = phi_a / (phi_b)  
    phi_inverse = M * (1 - np.exp(-quotient / M))
    return np.clip(phi_inverse, 0, M)

def scaler(img, constant, M):
    a = get_g(img,M)
    product = M - M*pow(1-(a/M), constant)
    return np.clip(product, 0, M)
```

```python
img1 = cv2.imread(r'Exercise Dataset\deadlift\deadlift_700001.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32) 
img2 = cv2.imread(r'Exercise Dataset\deadlift\deadlift_700021.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img3 = cv2.imread(r'Exercise Dataset\deadlift\deadlift_700041.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img4 = cv2.imread(r'Exercise Dataset\deadlift\deadlift_700091.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)
img5 = cv2.imread(r'Exercise Dataset\deadlift\deadlift_700101.jpg', cv2.IMREAD_GRAYSCALE).astype(np.float32)

img1 = cv2.resize(img1, (img3.shape[1], img3.shape[0]))
img2 = cv2.resize(img2, (img3.shape[1], img3.shape[0]))
img4 = cv2.resize(img4, (img3.shape[1], img3.shape[0]))
img5 = cv2.resize(img5, (img3.shape[1], img3.shape[0]))

plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.show()
```

```python
sum = log_add(img1, img2, 255)
dufference = log_subtract(img2, img3, 255)
product = log_multiply(img3, img4, 255)
quotient = log_divide(img4, img5, 255)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(sum, cmap='gray')
axes[0].set_title('Addition')
axes[0].axis('off')

axes[1].imshow(dufference, cmap='gray')
axes[1].set_title('Subtraction')
axes[1].axis('off')

axes[2].imshow(product, cmap='gray')
axes[2].set_title('Multiplication')
axes[2].axis('off')

axes[3].imshow(quotient, cmap='gray')
axes[3].set_title('Division')
axes[3].axis('off')
plt.show()

plt.tight_layout()
plt.show()
```
